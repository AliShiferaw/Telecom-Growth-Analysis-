```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = "C:/Users/alitr/OneDrive/Desktop/Ten Acadamey/Week Two/Data-20241218T114441Z-001/Data/Copy of Week2_challenge_data_source(CSV).csv"
df = pd.read_csv(file_path)

# Display the first few rows of the dataset
df.head()

# Check for any missing data or data types
df.info()

# Perform some simple visualizations
sns.countplot(data=df, x="Start")
plt.xticks(rotation=45)
plt.show()
```

```python
# Checking the percentage of missing values for each column
missing_data = df.isnull().mean() * 100
print(missing_data[missing_data > 0])
```

```python
import pandas as pd

# Assuming you have already loaded your dataset into a DataFrame 'df'

# Select only numeric columns
numeric_columns = df.select_dtypes(include=['number'])

# Impute missing values using the mean for each numeric column
df[numeric_columns.columns] = numeric_columns.apply(lambda col: col.fillna(col.mean()))

# Verify the missing values after imputation
missing_data_imputed = df.isnull().sum()
print("Missing data after imputation:")
print(missing_data_imputed[missing_data_imputed > 0])
```
